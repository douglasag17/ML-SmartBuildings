## Carga de datos "Historico Temp"

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
mainpath = "../Datasets/"
filename = "HISTORICO TEMP.xlsx"
fullpath = os.path.join(mainpath, filename)
```

```python
data = pd.read_excel(fullpath, "HISTORICO_TEMP")
```

## Se ordenan los datos en orden cronologico

```python
data = data.sort_values("Fecha- hora de lectura")
data
```

## Resumen estadístico básico de las variables numéricas
- count: numero de elementos no nulos
- mean: promedio (suma de valores / numero de elementos)
- std: desviacion estandar (que tanto se separan los valores del promedio)
- min: valor minimo
- 25%: primer cuartil
- 50%: segundo cuartil
- 75%: tercer cuartil
- max: valor maximo

```python
data.describe()
```

## La columna "TEMPERATURA RETORNO R14" tiene valores atipicos

```python
ejeX = data["Fecha- hora de lectura"]
ejeX = ejeX.to_numpy()
ejeY = data["TEMPERATURA RETORNO R14"]
ejeY = ejeY.to_numpy()
x = ejeX
y = ejeY
plt.xlabel("Fecha- hora de lectura")
plt.ylabel("TEMPERATURA RETORNO R14")
plt.scatter(x, y)
```

### Modificacion de los valores atipicos

```python
colR14 = data["TEMPERATURA RETORNO R14"]
colR14 = colR14.to_numpy()

for i in range(7102):
    if colR14[i] > 40:
        colR14[i] = 15.020601

colR14df = pd.DataFrame(colR14)
data['TEMPERATURA RETORNO R14'] = colR14df.values

data["TEMPERATURA RETORNO R14"].describe()
```

```python
ejeX = data["Fecha- hora de lectura"]
ejeX = ejeX.to_numpy()
ejeY = data["TEMPERATURA RETORNO R14"]
ejeY = ejeY.to_numpy()
x = ejeX
y = ejeY
plt.xlabel("Fecha- hora de lectura")
plt.ylabel("TEMPERATURA RETORNO R14")
plt.scatter(x, y)
```
